```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
images = np.load('total_three_datasets_sorted.npy')
```

```python
images.shape
```

```python
patient = 2
range(patient*38, (patient+1)*38)
```

```python
images_u_a = images[:1520,...]
```

```python
images_u_a.shape
```

```python
slices_u_a = [18, 19, 20]
slices_g = [30, 31, 32]
filtrado = np.empty([len(slices_u_a) * 60, 256, 256, 3])
```

Sabemos que hay 40 pacientes en images_u_a (20 de Utrecht y 20 de Amsterdam). Cada uno tiene 38 slices.

```python
capa = 0
for i in range(40):
    for valor in slices_u_a:
        filtrado[capa] = images[i*38 + valor, ...]
        capa += 1
```

El otro tiene 20 pacientes y 63 slices por paciente.

```python
capa = 120
for i in range(20):
    for valor in slices_g:
        filtrado[capa] = images[1520 + i*63 + valor, ...]
        capa += 1
```

```python
plt.imshow(filtrado[-3,..., 1], cmap='gray')
```

```python
np.save('muestra_seleccionada.npy', filtrado)
```
